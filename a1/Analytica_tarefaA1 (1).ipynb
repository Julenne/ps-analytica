```python
#1. Comece lendo o arquivo csv para um DataFrame.
data = pd.read_csv("dataset.csv")

df = pd.DataFrame(data)
#2. Visualize os dados em forma de tabela para compreender a estrutura e suas colunas.
df
```

```python
#3. Calcule o IDH medio do brasil no ano de 1991:
#(a) Comece fazendo uma filtragem nas linhas para selecionar apenas os dados
#referentes ao ano de 1991.
filtro = df["ano"] == 1991
df_filtrado = df[filtro]

df_filtrado
```

```python
#(b) Obtenha os valores de IDH.
idh = df_filtrado.iloc[:,5]
idh
```

```python
#(c) Retorne a media.
media = np.array(idh).mean()
print(media)
```

```python
#4. A partir da lista dos IDHs da questËœao anterior retorne os estados que estËœao
#acima da mÂ´edia, tambÂ´em no ano de 1991.
#(a) A partir do valor de mÂ´edia filtre o dataframe dos dados de 1991 para apenas
#os que possuem IDH maior que a mÂ´edia.
filtro2 = df["idhm"] > media
df_acimaMedia = df_filtrado[filtro2]
df_acimaMedia
```

```python
#(b) Retorne os estados referentes a essas linhas resultantes.
uf = df_acimaMedia.iloc[:,1]
uf
```

```python
#5. Ordene os dados do DataFrame de 1991 de acordo com o IDH, usando a funÂ¸cËœao
#sort values() e mostre a tabela dos 5 primeiros com o .head(5).
idh_crescente = df_filtrado.sort_values('idhm')
idh_crescente.head(5)
```

```python
#6. Analise os dados de IDH. Use as funÂ¸cËœ oes de .min(), .max(), .idxmin() e .idxmax()
#para saber qual o estado de maior e menor IDH no DataFrame do ano de 1991.
menor_idh = df_filtrado.iloc[:,5].idxmin()
maior_idh = df_filtrado.iloc[:,5].idxmax()
print("Estado com maior idh: ", df.iloc[maior_idh,1])
print("Estado com menor idh: ", df.iloc[menor_idh,1])
```

```python
# 2. Qual o nÃ­vel de desigualdade do Brasil?
filtro_91 = df["ano"] == 1991
df_1991 = df[filtro_91].iloc[:,5]

filtro_00 = df["ano"] == 2000
df_2000 = df[filtro_00].iloc[:,5]

filtro_10 = df["ano"] == 2010
df_2010 = df[filtro_10].iloc[:,5]

x = np.array(df_1991)
y = np.array(df_2000)
z = np.array(df_2010)

print("ANO 1991:")
plt.hist(x)
plt.show()
print("ANO 2000:")
plt.hist(y)
plt.show()
print("ANO 2010:")
plt.hist(z)
plt.show()

```

Observa-se que os dados estÃ£o bem prÃ³ximos um do outro e indo na direÃ§Ã£o de uma reta crescente, ou seja, existe relaÃ§Ã£o entre o IDH e a expectativa de vida. Quanto maior o IDH, maior a expectativa de vida.


```python
df_2010 = df[filtro_10]
df_1991 = df[filtro_91]
df_diferenca = df_2010.sort_values('sigla_uf')["expectativa_vida"].reset_index() - df_1991.sort_values('sigla_uf')["expectativa_vida"].reset_index()

x = np.array(df_2010["sigla_uf"])
y = np.array(df_diferenca.iloc[:,1])


plt.figure(figsize=(10,6))
plt.bar(x,y)
plt.show()
```

```python

df_dif_filtrado = df_diferenca[ df_diferenca.iloc[:,1] >= 10 ]

all_ufs = []
for i in df_dif_filtrado.index:
  all_ufs.append(df_2010.sort_values('sigla_uf')["sigla_uf"].iloc[i])

x = np.array(all_ufs)
y = np.array(df_dif_filtrado.iloc[:,1])

plt.bar(x,y)
plt.show()
```
